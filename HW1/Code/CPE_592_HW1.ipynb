```python
import PIL.Image as Image

ENCODED_BIT = 1 - 1

# 1. Convert the string to binary
string = "Christopher Spadavecchia"
binary = [bin(ord(char))[2:].zfill(8) for char in string]
binary.append("00000000")
binary_string = "".join(binary)

# 2. Load the image
image = Image.open("input.jpg")
pixels = list(image.getdata())

# 3. Encode the binary string into the image using the least significant bit of each color channel (RGB)
for i in range(len(binary_string)):
    index = 250000 + i // 3
    pixel_index = i % 3
    
    pixel = list(pixels[index])
    if binary_string[i] == "1":
        pixel[pixel_index] |= (1 << ENCODED_BIT)
    else:
        pixel[pixel_index] &= ~(1 << ENCODED_BIT)

    pixels[index] = tuple(pixel)




# 4. Save the image
image.putdata(pixels)
image.save("output.png")

# 5. Decode the image
binary = []
index = 250000
digit = ""
while True:
    for i in range(3):
        pixel = pixels[index]
        print(pixel)
        digit += bin(pixel[i])[2:].zfill(8)[7 - ENCODED_BIT]
        if len(digit) == 8:
            if digit == "00000000":
                break
            else:
                binary.append(digit)
                digit = ""

    index += 1
    if digit == "00000000":
        break

# 6. Convert the binary string to a string
string = "".join([chr(int(byte, 2)) for byte in binary])
print(string)
```
